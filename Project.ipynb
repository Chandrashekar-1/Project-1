```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import ast  # for parsing genre strings
from sklearn.preprocessing import MultiLabelBinarizer, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the mobile sales data csv file
df = pd.read_csv("mobile sales data.csv", low_memory=False)
```

```python
#checking all the columns from datasets
df.columns
```

```python
mobiles = df[['Customer Name', 'Product', 'Brand', 'Price', 'Processor Specification', 'RAM']].copy()
```

```python
mobiles = mobiles.dropna(subset=['Customer Name', 'Brand', 'Price', 'RAM'])
```

```python
def parse_product(product_str):
    """Converts a string of genre dictionaries into a list of genre names."""
    try:
        product = ast.literal_eval(product_str)
        return [p['name'] for p in product]
    except:
        return []

# Apply the parsing function to each row
mobiles['Product'] = mobiles['Product'].apply(parse_product)
```

```python
# Count the occurrences of each brand
Brand_counts = mobiles['Brand'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=Brand_counts.index, y=Brand_counts.values, palette='viridis')
plt.title('Brand Distribution')
plt.xticks(rotation=45)
plt.ylabel('Number of Mobiles')
plt.xlabel('Brand')
plt.tight_layout()
plt.show()
```

```python
# Group by Brand and sum the Quantity Sold
brand_quantity = df.groupby('Brand')['Quantity Sold'].sum().sort_values(ascending=False)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_quantity.index, y=brand_quantity.values, palette='coolwarm')
plt.title('Quantity Sold by Brand')
plt.xticks(rotation=45)
plt.ylabel('Total Quantity Sold')
plt.xlabel('Brand')
plt.tight_layout()
plt.show()
```

```python
# Create a table for the top 10 brands and their prices
top_10_brands = mobiles.groupby('Brand')['Price'].mean().astype(int).sort_values(ascending=False).head(10)
top_10_table = top_10_brands.reset_index()
top_10_table.columns = ['Brand', 'Price']
print(top_10_table)
```
